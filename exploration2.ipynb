```python
## load the df
import os, sys
import pandas as pd
from globes import taxi_dir, days_dir, zoneIdToName

## build a humungous df from csvs
dfs = []
for filename in os.listdir(os.path.join(taxi_dir, days_dir)):
    if filename.endswith(".csv") and "01-08" in filename:
        filepath = os.path.join(taxi_dir, days_dir, filename)
        df = pd.read_csv(filepath, parse_dates=['pickup_datetime', 'dropoff_datetime'], index_col=0)
        dfs.append(df)

df_all = pd.concat(dfs)
df_all
```

```python
## extract features from df
X = df[['passenger_count', 'pickup_zone_taxi', 'dropoff_zone_taxi']]

## get rid of SettingWithCopyWarning
X.is_copy = False
df.is_copy = False

X['dropoff_zone_taxi'] = X['dropoff_zone_taxi'].fillna(-1).astype("int")
X['pickup_zone_taxi'] = X['pickup_zone_taxi'].fillna(-1).astype("int")
X['pickup_day'] = df['pickup_datetime'].apply(lambda t: t.weekday())
X['pickup_hour'] = df['pickup_datetime'].apply(lambda t: t.hour)
X
```

```python
zone_data = zoneIdToName()
```

```python
## for each pickup zone, group by dropoff_zone and count
def get_counts_as_row(pickup_zone, group):
    x = pd.DataFrame(group.groupby('dropoff_zone_taxi', as_index=False).size().rename('count'))
    x = x.reset_index(level=0)
    x['pickup_zone'] = pickup_zone
    x = x.pivot(index='pickup_zone', columns='dropoff_zone_taxi', values='count')
    return x
    

## group by pickup zone
grouped = X.groupby('pickup_zone_taxi', as_index=False)
rows = []
for name, group in grouped:
    x = get_counts_as_row(name, group)
    rows.append(x)
    
dests = pd.concat(rows).fillna(0)
print dests.index
print dests
    



# most common pickup zone?
def top_pickup_zone():                       
    grouped = pd.DataFrame(X.groupby('pickup_zone_taxi', as_index=False).size().rename('counts'))
    grouped = grouped.sort_values(by="counts", ascending=False)
    topZone = grouped.iloc[[10]].index[0]
    # topZone = 1
    print "using zone: " + zone_data[topZone]
```

```python
""" Given df, return dataframe with dest counts for each day, hour, pickup zone
"""
def _get_dests(X):

    day_groups = X.groupby('pickup_day');
    for day, day_group in day_groups:
        hour_groups = day_group.groupby('pickup_hour')
        for hour, hour_group in hour_groups:
            grouped = hour_group.groupby('pickup_zone_taxi', as_index=False)
            rows = []
            for pickup_zone, group in grouped:
                x = pd.DataFrame(group.groupby('dropoff_zone_taxi', as_index=False).size().rename('count'))
                x = x.reset_index(level=0)
                x['pickup_zone'] = pickup_zone
                
                x = x.pivot(index='pickup_zone', columns='dropoff_zone_taxi', values='count')
                x['pickup_zone'] = x.index
                x['pickup_day'] = day
                x['pickup_hour'] = hour
                rows.append(x)

    dests = pd.concat(rows).fillna(0)
    dests = dests.set_index(['pickup_day', 'pickup_hour', 'pickup_zone'])
    return dests

_get_dests(X) + _get_dests(X)
```
