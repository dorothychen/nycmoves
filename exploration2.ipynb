```python
## load the df

import os, sys
import pandas as pd
from globes import taxi_dir, days_dir

filename = "zones_2016-01-01.csv"
filepath = os.path.join(taxi_dir, days_dir, filename)

df = pd.read_csv(filepath, parse_dates=['pickup_datetime', 'dropoff_datetime'], index_col=0)
df
```

```python
## extract features from df
X = df[['passenger_count', 'pickup_zone_taxi', 'dropoff_zone_taxi']]
X['dropoff_zone_taxi'] = X['dropoff_zone_taxi'].fillna(-1).astype("int")
X['pickup_zone_taxi'] = X['pickup_zone_taxi'].fillna(-1).astype("int")
X['pickup_day'] = df['pickup_datetime'].apply(lambda t: t.weekday())
X['pickup_hour'] = df['pickup_datetime'].apply(lambda t: t.hour)
X
```

```python
import json
ZONE = "taxi"
ID = "LocationID"
NAME = "zone"

def getGeoJSON():
    zone_filename = "zones_" + ZONE + ".geojson"
    with open(zone_filename) as zone_file:
            zone_geojson = json.load(zone_file)
            zone_geojson = map(lambda x: x["properties"], zone_geojson["features"])
    zone_data = {-1: "None"}
    for zone in zone_geojson:
        zone_data[zone[ID]] = zone[NAME]
    return zone_data

zone_data = getGeoJSON()
```

```python
# most common pickup zone?
grouped = pd.DataFrame(X.groupby('pickup_zone_taxi', as_index=False).size().rename('counts'))
grouped = grouped.sort_values(by="counts", ascending=False)
topZone = grouped.iloc[[0]].index[0]
# topZone = 1
print "using zone: " + zone_data[topZone]

# what are the most common dropoff zones?
X2 = X.loc[X['pickup_zone_taxi'] == topZone]
dests = pd.DataFrame(X2.groupby('dropoff_zone_taxi', as_index=False).size().rename('counts'))
dests = dests.sort_values(by="counts", ascending=False)
dests['zone'] = dests.index
dests['zone_name'] = dests['zone'].apply(lambda x: zone_data[x])
print 'most frequent destination: ' + dests.iloc[0]['zone_name']
```

```python
# make histogram of destinations
import matplotlib.pyplot as plt

data = dests['counts']
binwidth = 20
plt.hist(data, bins=range(min(data), max(data) + binwidth, binwidth))
plt.show()
```
